```python
import pandas as pd
import numpy as np
from scipy.io import loadmat 
```

The data comes in as a .mat file from matlab, and the structure is a little wierd, and not well documented.  The get_labels functions creates a dataframe with the filnames and meta-data and writes it out to a csv for future use.

```python
X = loadmat('/storage/stanford_cars/cars_annos.mat')
```

```python
def get_labels():
    MAT = loadmat('/storage/stanford_cars/cars_annos.mat')
    annotations = MAT["annotations"][0,:]
    nclasses = len(MAT["class_names"][0])
    class_names = dict(zip(range(nclasses),[c[0] for c in MAT["class_names"][0]]))
    labelled_images = {}
    index = []
    dataset = []
    for arr in annotations:
        # the first entry in the row is the image name
        index.append(arr[0][0])
        # The rest is the data, first bbox, then classid then a boolean for whether in train or test set
        dataset.append([y[0][0] for y in arr][1:])
    # Convert to a DataFrame, and specify the column names
    DF = pd.DataFrame(dataset, index, columns =["BBOX_Y2","BBOX_X1","BBOX_Y1","BBOX_X2","ClassID","TestSet"])
    DF = DF.assign(ClassName=DF.ClassID.map(dict(class_names)))
    return DF

labelDF = get_labels()
```

Write the data out to a csv file.

```python
labelDF.reset_index().to_csv('/storage/stanford_cars/cars_annos.csv', index=False)
```
